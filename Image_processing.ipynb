**Conversion of color image (Rose.jpg and Lenna.tif)  to grayscale image using  a def function and list of comprehension method and plot processing time comparison with  OpenCV and for loop methods**

 Solution:
 Converting lenna.tif image to greyscale using the formula:
 Greyscale value at (i,j) = 0.2989 * r(i,j) + 0.5870 * g(i,j) + 0.1140 * b(i,j)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time 


def toGreyscaleForloop(img):
    dim = img.shape
    for i in range(dim[0]):
        for j in range(dim[1]):
            img[i][j] = (0.2989 * img[i][j][2]) + (0.587 * img[i][j][1]) + (0.114 * img[i][j][0])
    return img


def toGreyscaleListComp(img):
    dim = img.shape
    img = [[(0.2989 * img[i][j][2]) + (0.587 * img[i][j][1]) + (0.114 * img[i][j][0]) for j in range(dim[1])] for i in range(dim[0])]
    return img


#Nested for loop method
img = cv2.imread('lenna.tif')
plt.imshow(img)
plt.title('Input image in BGR format')
plt.show()

t1 = time.time()
img = toGreyscaleForloop(img)
t2 = time.time()

plt.imshow(img)
plt.title('Output Greyscale image using for loop')
plt.show()

dt1 = t2 - t1
print('Time taken for nested for loop: ')
print(dt1)


#Inbuilt OpenCV2 method
img = cv2.imread('lenna.tif')

t1 = time.time()
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
t2 = time.time()

plt.imshow(img2, cmap = 'gray')
plt.title('Output Greyscale image using BGR2GRAY')
plt.show()

dt2 = t2 - t1
print('Time taken for cv2 inbuilt function: ')
print(dt2)


#Nested list comprehension method
img = cv2.imread('lenna.tif')

t1 = time.time()
img = toGreyscaleListComp(img)
t2 = time.time()

plt.imshow(img2, cmap = 'gray')
plt.title('Output Greyscale image using List of comprehension')
plt.show()

dt3 = t2 - t1
print('Time taken for cv2 List of comprehension method: ')
print(dt3)


#Plotting the comparision between processing times
y = [dt1, dt2, dt3]
x_label = ['For loop', 'BGR2GRAY', 'List comprehension']
x = [i for i in range(len(x_label))]

plt.bar(x,y)
plt.xlabel('Method used ->')
plt.ylabel('Runtime ->')
plt.title('Comparision of Processing time for different methods')
plt.xticks(x, x_label)
plt.show()
```

The bar plot show the comparision of processing time for nexted for loop, nested list of comprehension and inbuilt OpenCV2 method.

**Conversion of Grayscale  image to Binary images using a def function and list of comprehension method and  plot processing time comparison with OpenCV and for loop methods**

Solution:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time


def grayToBinaryForLoop(img):
    dim = img.shape
    for i in range(dim[0]):
        for j in range(dim[1]):
            if img[i][j] <= 127:
                img[i][j] = 0
            else:
                img[i][j] = 255
    return img


def grayToBinaryListComp(img):
    dim = img.shape
    img = [[0 if img[i][j] <= 127 else 255 for j in range(dim[1])] for i in range(dim[0])]
    return img


#Nested for loop method
img = cv2.imread('lenna.tif', 0)
plt.imshow(img, cmap='gray')
plt.title('Input greyscale image')
plt.show()

t10 = time.time()
img = grayToBinaryForLoop(img)
t11 = time.time()

plt.imshow(img, cmap='gray')
plt.title('Ouput Binary image using for loop')
plt.show()

dt1 = t11 - t10
print('Processing time while using for loop: ')
print(dt1)


#Nested list comprehension method
img = cv2.imread('lenna.tif', 0)

t20 = time.time()
img = grayToBinaryListComp(img)
t21 = time.time()

plt.imshow(img, cmap='gray')
plt.title('Ouput Binary image using for loop')
plt.show()

dt2 = t21 - t20
print('Processing time while using for loop: ')
print(dt2)

#CV2 inbuilt method
img = cv2.imread('lenna.tif', 0)

t30 = time.time()
(thresh, BnWImg) = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
t31 = time.time()

plt.imshow(BnWImg, cmap='gray')
plt.title('Ouput Binary image using for loop')
plt.show()

dt3 = t31 - t30
print('Processing time while using for loop: ')
print(dt3)


#Plotting the comparision between processing times
y = [dt1, dt2, dt3]
x_label = ['For loop', 'List comprehension', 'cv2.threshold']
x = [i for i in range(len(x_label))]

plt.bar(x,y)
plt.xlabel('Method used ->')
plt.ylabel('Runtime ->')
plt.title('Comparision of Processing time for different methods')
plt.xticks(x, x_label)
plt.show()
```

**Add two images (image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and for loop methods.**

Solution:

```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
import matplotlib.image as image


def add2ImagesFL(img1, img2):
    dim = img1.shape
    t1 = time.time()
    for i in range(dim[0]):
        for j in range(dim[1]):
            for c in range(dim[2]):
                temp = int(img1[i][j][c]) + int(img2[i][j][c])
                if (temp > 255):
                    img[i][j][c] = 255
                else:
                    img[i][j][c] = temp;
    t2 = time.time()
    dt = t2 - t1
    return (img, dt)


def add2ImagesLC(img1, img2):
    dim = img1.shape
    t1 = time.time()
    img = [[[int(img1[i][j][c]) + int(img2[i][j][c]) for c in range(dim[2])] for j in range(dim[1])] for i in range(dim[0])]
    t2 = time.time()
    dt = t2 - t1
    return (img, dt)


#Input images
img1 = cv2.imread('input1.jpg')
plt.imshow(img1)
plt.title('Image1')
plt.show()

img2 = cv2.imread('input2.jpg')
plt.imshow(img2)
plt.title('Image2')
plt.show()


#For loop
(imgA, dt1) = add2ImagesFL(img1, img2)
plt.imshow(imgA)
plt.title('Image Addition using for loop')
plt.show()
print(dt1)


#List of comprehension
(imgB, dt2) = add2ImagesLC(img1, img2)
plt.imshow(imgB)
plt.title('Image Addition using list comprehension')
plt.show()
print(dt2)


#CV2 inbuilt function
t1 = time.time()
imgC = cv2.add(img1, img2)
t2 = time.time()
dt3 = t2 - t1
plt.imshow(imgC)
plt.title('Image Addition using cv2.add function')
plt.show()
print(dt3)


#Plotting the comparision of processing times
x = [1, 2, 3]
y = [dt1, dt2, dt3]
x_label = ['For loop', 'List comp', 'cv2.add']
plt.bar(x,y)
plt.xlabel('Method used ->')
plt.ylabel('Runtime ->')
plt.title('Comparision of Processing time for different methods')
plt.xticks(x, x_label)
plt.show()
```

**Subtraction of two images (image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and for loop methods.**

Solution:

```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
import matplotlib.image as image


def add2ImagesFL(img1, img2):
    dim = img1.shape
    t1 = time.time()
    for i in range(dim[0]):
        for j in range(dim[1]):
            for c in range(dim[2]):
                img[i][j][c] = abs(int(img1[i][j][c]) - int(img2[i][j][c]))
    t2 = time.time()
    dt = t2 - t1
    return (img, dt)


def add2ImagesLC(img1, img2):
    dim = img1.shape
    t1 = time.time()
    img = [[[abs(int(img1[i][j][c]) - int(img2[i][j][c])) for c in range(dim[2])] for j in range(dim[1])] for i in range(dim[0])]
    t2 = time.time()
    dt = t2 - t1
    return (img, dt)


#Input images
img1 = cv2.imread('input1.jpg')
plt.imshow(img1)
plt.title('Image1')
plt.show()

img2 = cv2.imread('input2.jpg')
plt.imshow(img2)
plt.title('Image2')
plt.show()


#For loop
(imgA, dt1) = add2ImagesFL(img1, img2)
plt.imshow(imgA)
plt.title('Image Subtraction using for loop')
plt.show()
print(dt1)


#List of comprehension
(imgB, dt2) = add2ImagesLC(img1, img2)
plt.imshow(imgB)
plt.title('Image Subtraction using list comprehension')
plt.show()
print(dt2)


#CV2 inbuilt function
t1 = time.time()
imgC = cv2.subtract(img1, img2)
t2 = time.time()
dt3 = t2 - t1
plt.imshow(imgC)
plt.title('Image Subtraction using cv2.add function')
plt.show()
print(dt3)


#Plotting the comparision of processing times
x = [1, 2, 3]
y = [dt1, dt2, dt3]
x_label = ['For loop', 'List comp', 'cv2.add']
plt.bar(x,y)
plt.xlabel('Method used ->')
plt.ylabel('Runtime ->')
plt.title('Comparision of Processing time for different methods')
plt.xticks(x, x_label)
plt.show()
```
